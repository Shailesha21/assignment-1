```python
Q6. How do conditional statements work in Python? Illustrate with examples

ans:- -Conditional statements are used to make decisions in a Python program.
-They help the program choose different actions based on different conditions.
-They use if, elif (else if), and else keywords.

1.if Statement
Checks if a condition is true and runs the code block under it.

Example:

age = 18
if age >= 18:
    print("You are an adult.")
    
2. if-else Statement
Runs one block if the condition is true, and another if it is false.

Example:

marks = 40
if marks >= 50:
    print("You passed.")
else:
    print("You failed.")
    
3. if-elif-else Statement
Checks multiple conditions one by one.
Runs the block of the first true condition.

Example:

score = 75

if score >= 90:
    print("Grade A")
elif score >= 70:
    print("Grade B")
elif score >= 50:
    print("Grade C")
else:
    print("Grade F")
    
-How it Works Internally
Python checks conditions from top to bottom.
As soon as it finds a true condition, it runs that block and skips the rest.

-:Important Points
Indentation (spacing) is very important.
You can use logical operators like and, or, not in conditions.

Example using and:

age = 25
income = 30000

if age > 18 and income > 20000:
    print("Eligible for loan")
```
